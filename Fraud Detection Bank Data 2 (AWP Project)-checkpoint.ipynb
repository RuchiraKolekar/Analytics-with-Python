## Data Cleaning

```python
df.isnull().sum()
```

* isnull().sum()- Shows the number of null values present in the data as per each column in the data.
* To handle the missing values there are 3 rules
* If the missing values are less than 2-3% than we use drop na.
* If the missing values are between 4-50% than will fill na.
* If missing values are more than 50% than we drop that column.

```python
df.dropna(inplace=True)
```

* df.dropna()- Helps us to drop the null records present in our data so that when we carry out analysis it does not go wrong.

```python
df.isnull().sum()
```

* After doing the above step of dropna, we are rechecking wether we still have null values in our data by doing isnull.
